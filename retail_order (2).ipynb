```python
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename columns names ..make them lower case and replace space with underscore
#METHOD 1 :
df.rename(columns={'Order Id':'order_id', 'City':'city'})
```

```python
#Method 2:
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
#Replacing (space) to (-):
df.columns=df.columns.str.replace(' ','_')
df.head(5)
```

```python
#CALCULATING DISCOUNTS---derive new columns discount , sale price and profit
df['list_price']*df['discount_percent']*.01
```

```python
df['discount']=df['list_price']*df['discount_percent']*.01
```

```python
#SALES :
df['sale_price']= df['list_price']-df['discount']
```

```python
#PROFIT
df['profit']=df['sale_price']-df['cost_price']
```

```python
import pymysql
```

```python
# Database connection parameters
db_config = {
    'host': 'localhost',
    'user': 'root',
    'password': 'root'
}
```

```python
conn = pymysql.connect(**db_config)
cursor = conn.cursor()
```

```python
cursor.execute("Create database if not exists retail_orders")
```

```python
cursor.execute("Use retail_orders")
```

```python
cursor.execute("Create table if not exists df_orders(order_id int, order_date date, ship_mode varchar(20), segment varchar(20),country varchar(20), city varchar(20), state varchar(20), postal_code varchar(20), region varchar(20), category varchar(20), sub_category varchar(20), product_id varchar(50), quantity int, discount decimal(7,2), sale_price decimal(7,2), profit decimal(7,2))")
               
```

```python
query = "INSERT INTO df_orders (order_id,order_date,ship_mode,segment,country, city,state,postal_code, region, category, sub_category,product_id, quantity, discount, sale_price, profit) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
```

```python
# Convert DataFrame to list of lists
data_list = df.values.tolist()
data_list
```

```python
# Execute the insert query for each row in the DataFrame
for row in data_list:
    cursor.execute(query, row)
```

```python
# Commit the transaction
conn.commit()
```

# Questions

```python
# Find top 10 highest revenue generating products

cursor.execute('''select product_id,sum(sale_price) as sales
from df_orders
group by product_id
order by sales desc
limit 10;''')
table1 = cursor.fetchall()
answer_1 = pd.DataFrame(table1,columns=["product_id","sales"])
answer_1
```

```python
# Find the top 5 cities with the highest profit margins

cursor.execute("""
               select city, sum(profit) as total_profit 
               from df_orders 
               group by city 
               order by total_profit desc 
               limit 5;
               """)
table2 = cursor.fetchall()
answer_2 = pd.DataFrame(table2, columns =["city", "total_profit"])
answer_2
```

```python
# Ensure there are no unread results from previous queries
while cursor.nextset():
    pass
```

```python
# Calculate the total discount given for each category

cursor.execute("""select category, sum(discount) as total_discount
from df_orders
group by category;""")

table3 = cursor.fetchall()

answer_3 = pd.DataFrame(table3,columns=["category","total_discount"])
answer_3
```

```python
# Find the average sale price per product category

cursor.execute("""select category, avg(sale_price) as avg_sale_price
from df_orders
group by category;""")

table4 = cursor.fetchall()

answer_4 = pd.DataFrame(table4,columns=["category","avg_sale_price"])
answer_4
```

```python
# Determine the most popular ship mode used for orders

cursor.execute("""select ship_mode, count(*) as count_orders
from df_orders
group by ship_mode
order by count_orders desc;""")

table5 = cursor.fetchall()

answer_5 = pd.DataFrame(table5,columns=["ship_mode","count_orders"])
answer_5
```

```python
# List the states with the highest number of orders

cursor.execute("""select state, count(order_id) as number_of_orders
from df_orders
group by state
order by number_of_orders desc;""")

table6 = cursor.fetchall()

answer_6 = pd.DataFrame(table6,columns=["state","number_of_orders"])
answer_6
```

```python
# Calculate the total revenue generated per year

cursor.execute("""select year(order_date) as order_year, sum(sale_price) as total_revenue
from df_orders
group by year(order_date)
order by order_year;""")

table7 = cursor.fetchall()

answer_7 = pd.DataFrame(table7,columns=["order_year","total_revenue"])
answer_7
```

```python
# Find the product category with the highest total profit

cursor.execute("""select category, sum(profit) as total_profit
from df_orders
group by category
order by total_profit desc;""")

table8 = cursor.fetchall()

answer_8 = pd.DataFrame(table8,columns=["category","total_profit"])
answer_8
```

```python
# Determine the average discount percentage given per region

cursor.execute("""select region, 
       avg((discount / (sale_price + discount)) * 100) as avg_discount_percent
from df_orders
group by region;
;""")

table9 = cursor.fetchall()

answer_9 = pd.DataFrame(table9,columns=["region","avg_discount_percent"])
answer_9
```

```python
# Identify the top 3 segments with the highest quantity of orders

cursor.execute("""select segment, sum(quantity) as total_quantity
from df_orders
group by segment
order by total_quantity desc
limit 3;""")

table10 = cursor.fetchall()

answer_10 = pd.DataFrame(table10,columns=["segment","total_quantity"])
answer_10
```

```python
# Find the total profit per category

cursor.execute("""select category, sum(profit) as total_profit
from df_orders
group by category
order by total_profit desc;""")

table11 = cursor.fetchall()

answer_11 = pd.DataFrame(table11,columns=["category","total_profit"])
answer_11
```

```python
# Find the region with the highest average sale price

cursor.execute("""select region, avg(sale_price) as avg_sale_price
from df_orders
group by region
order by avg_sale_price desc;""")

table12 = cursor.fetchall()

answer_12 = pd.DataFrame(table12,columns=["region","avg_sale_price"])
answer_12
```
